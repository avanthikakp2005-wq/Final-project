```python
plt.figure()
df["AgeGroup"].value_counts().plot(kind="bar")
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
plt.figure()
df["SM_Hours"].value_counts().plot(kind="bar")
plt.title("Social Media Usage Hours")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
df["SM_Hours"].value_counts().plot(kind="bar")
plt.title("Social Media Usage Hours")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
age_hours = pd.crosstab(df["AgeGroup"], df["SM_Hours"])
print(age_hours)

plt.figure()
age_hours.plot(kind="bar")
plt.title("Age Group vs Social Media Hours")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
gender_productivity = pd.crosstab(df["Gender"], df["ProductivityImpact"])
print(gender_productivity)

plt.figure()
gender_productivity.plot(kind="bar")
plt.title("Gender vs Productivity Impact")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
platform_counts = df["MostUsedPlatform"].value_counts()

plt.figure()
plt.pie(platform_counts, labels=platform_counts.index, autopct="%1.1f%%")
plt.title("Most Used Social Media Platforms")

plt.show()
```

```python
plt.figure()
plt.hist(df["SM_Hours"])
plt.title("Distribution of Social Media Hours")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show()
```

```python
encoded_df = df.copy()
for col in encoded_df.columns:
    encoded_df[col] = encoded_df[col].astype("category").cat.codes

corr = encoded_df.corr()
print(corr)

plt.figure()
plt.imshow(corr)
plt.title("Correlation Heatmap")
plt.colorbar()
plt.show()
```
